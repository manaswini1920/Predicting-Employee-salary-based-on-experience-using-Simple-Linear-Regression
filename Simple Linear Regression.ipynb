```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import dataset
dataset=pd.read_csv("C:/Users/manas/Desktop/Salary_Data.csv")
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
#splitting dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#training simple linear regression model on training set
from sklearn.linear_model import LinearRegression
regre = LinearRegression()
regre.fit(x_train,y_train)
```

```python
#predicting the test set 
y_pred=regre.predict(x_test)
```

```python
#visualising training set results
plt.scatter(x_train,y_train,color='red')
#plotting predicting the salaries of train set with exp
plt.plot(x_train,regre.predict(x_train),color='blue')
#title
plt.title("Experience vs Salary(Training Set)")
plt.xlabel("years of Training")
plt.ylabel("Salary")
plt.show()



```

```python
#visualising testing set results
plt.scatter(x_test,y_test,color='red')
#plotting predicting the salaries of test/train set with exp
#giving any test/train as input is okay as both will have same best fit line
plt.plot(x_train,regre.predict(x_train),color='blue')
#title
plt.title("Experience vs Salary(Test Set)")
plt.xlabel("years of Training")
plt.ylabel("Salary")
plt.show()

```
